```python
# Importações necessárias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

from datetime import datetime
```

```python
# Carregar os dados
customers_df = pd.read_csv('../data/olist_customers_dataset.csv')
geolocation_df = pd.read_csv('../data/olist_geolocation_dataset.csv')
orders_items_df = pd.read_csv('../data/olist_order_items_dataset.csv')
payments_df = pd.read_csv('../data/olist_order_payments_dataset.csv')
reviews_df = pd.read_csv('../data/olist_order_reviews_dataset.csv')
orders_df = pd.read_csv('../data/olist_orders_dataset.csv')
products_df = pd.read_csv('../data/olist_products_dataset.csv')
sellers_df = pd.read_csv('../data/olist_sellers_dataset.csv')




print(customers_df.head())
print(geolocation_df.head())
print(orders_items_df.head())
print(payments_df.head())
print(reviews_df.head())
print(orders_df.head())
print(products_df.head())
print(sellers_df.head())

```

```python
# Fazer o merge entre os datasets
merged_df = pd.merge(orders_items_df, products_df, on='product_id', how='inner')

# Calcular o preço médio por categoria
avg_price_by_category = merged_df.groupby('product_category_name')['price'].mean().reset_index()

# Plotando o gráfico de barras para o preço médio por categoria de produto
plt.figure(figsize=(12,8))
sns.barplot(x='product_category_name', y='price', data=avg_price_by_category, palette="viridis")

# Ajustando o gráfico para melhorar a leitura
plt.title("Preço médio por categoria de produto")
plt.xlabel("Categoria do produto")
plt.ylabel("Preço médio (R$)")
plt.xticks(rotation=90)  # Rotacionar os nomes das categorias para melhor visualização
plt.show()
```

```python
# Fazer o merge entre os datasets
merged_df = pd.merge(orders_items_df, reviews_df, on='order_id', how='inner')

sns.scatterplot(x='price', y='review_score', data=merged_df)
plt.title("Preço vs Avaliação do produto")
plt.xlabel("Preço")
plt.ylabel("Avaliação")
plt.show()
```

```python
# Fazer o merge entre os datasets
merged_df = pd.merge(orders_df, customers_df, on='customer_id', how='inner')
merged_df = pd.merge(merged_df, orders_items_df, on='order_id', how='inner')

sales_by_state = merged_df.groupby('customer_state')['price'].sum().sort_values(ascending=False)
sns.barplot(x=sales_by_state.index, y=sales_by_state.values, palette='viridis')
plt.title("Vendas totais por estado")
plt.xlabel("Estado")
plt.ylabel("Vendas totais (R$)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Fazer o merge entre os datasets
merged_df = pd.merge(orders_items_df, products_df, on='product_id', how='inner')
# merged_df = pd.merge(merged_df, reviews_df, on='order_id', how='inner')


sns.scatterplot(x='product_photos_qty', y='price', data=merged_df)
plt.title("Preço médio vs. Quantidade de fotos do produto")
plt.xlabel("Quantidade de fotos")
plt.ylabel("Preço")
plt.show()
```

```python
# Fazer o merge entre os datasets
df = pd.merge(orders_items_df, orders_df, on='order_id', how='inner')


df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['month'] = df['order_purchase_timestamp'].dt.month
monthly_sales = df.groupby('month')['price'].sum()
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, color='green')
plt.title("Vendas por mês")
plt.xlabel("Mês")
plt.ylabel("Vendas totais em milhões (R$)")
plt.show()
```

```python
sns.scatterplot(x='freight_value', y='price', data=orders_items_df)
plt.title("Relação entre preço e valor do frete")
plt.xlabel("Valor do frete (R$)")
plt.ylabel("Preço (R$)")
plt.show()
```

```python
# merge entre orders_items_df e payments_df
df = pd.merge(orders_items_df, payments_df, on='order_id', how='inner')

# valores em absoluto
payment_sales = df.groupby('payment_type')['price'].sum().sort_values(ascending=False)
sns.barplot(x=payment_sales.index, y=payment_sales.values, palette="coolwarm")
plt.title("Vendas por tipo de pagamento")
plt.xlabel("Tipo de pagamento")
plt.ylabel("Vendas totais em 10 milhões (R$)")
plt.show()
```

```python
# merge entre orders_df e orders_items_df
df = pd.merge(orders_df, orders_items_df, on='order_id', how='inner')

# Garantir que as colunas de data estão no formato datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])

# Calcular o tempo de entrega (diferença entre a data de entrega e a data de compra)
df['delivery_time'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# Plotar a relação entre o valor do frete e o tempo de entrega
sns.scatterplot(x='freight_value', y='delivery_time', data=df)
plt.title("Relação entre frete e tempo de entrega")
plt.xlabel("Valor do frete (R$)")
plt.ylabel("Tempo de entrega (dias)")
plt.show()
```

```python
# fazendo os merge entre orders_items_df e customers_df
df = pd.merge(orders_items_df, orders_df, on='order_id', how='inner')
df = pd.merge(df, customers_df, on='customer_id', how='inner')


freight_by_state = df.groupby('customer_state')['freight_value'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=freight_by_state.index, y=freight_by_state.values, palette="magma")
plt.title("Frete médio por estado")
plt.xlabel("Estado")
plt.ylabel("Frete médio (R$)")
plt.show()
```

```python
# Fazer o merge entre os datasets
merged_df = pd.merge(orders_items_df, products_df, on='product_id', how='inner')
merged_df = pd.merge(merged_df, orders_df, on='order_id', how='inner')

# criar uma nova coluna: tempo de entrega
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
merged_df['order_delivered_customer_date'] = pd.to_datetime(merged_df['order_delivered_customer_date'])
merged_df['delivery_time'] = (merged_df['order_delivered_customer_date'] - merged_df['order_purchase_timestamp']).dt.days


# Selecionar apenas colunas numéricas relevantes
correlation_df = merged_df[['price', 'freight_value', 'product_photos_qty',
                            'product_weight_g', 'product_length_cm', 
                            'product_height_cm', 'product_width_cm', 'delivery_time']]

# Criar a matriz de correlação
corr_matrix = correlation_df.corr()

# Plotar o heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Matriz de Correlação entre Variáveis Numéricas")
plt.show()
```
